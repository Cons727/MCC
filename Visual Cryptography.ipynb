```python
from PIL import Image
import numpy as np
saved = "t_image.jpeg"
size = 200,200
im =  Image.open("mew.jpeg")
im.thumbnail(size)
im.save(saved)
[t_r, t_g, t_b] = im.split()


red_bin = [[0 for x in range(200)] for x in range(200)]
green_bin = [[0 for x in range(200)] for x in range(200)]
blue_bin = [[0 for x in range(200)] for x in range(200)] 

for i in range(200):
    for j in range(200):
        cordinate = x, y = j, i
        red_bin[i][j] = bin(t_r.getpixel(cordinate))[2:].zfill(8)
        green_bin[i][j] = bin(t_g.getpixel(cordinate))[2:].zfill(8)
        blue_bin[i][j] = bin(t_b.getpixel(cordinate))[2:].zfill(8)
```
